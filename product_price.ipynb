```python
# import dependences 

import pandas as pd
from sqlalchemy import create_engine, inspect
import numpy as np
import datetime
import psycopg2
```

```python
#read in csv's

csv_file = "Resources/items.csv"
items = pd.read_csv(csv_file)
#items.set_index('id', inplace=True)
items
```

```python
csv_file = "Resources/lk_store.csv"
stores = pd.read_csv(csv_file)
#stores.set_index('id', inplace=True)
stores
```

```python
csv_file = "Resources/web_scrape_data.csv"
scraped_data = pd.read_csv(csv_file)
#scraped_data.set_index('product_name', inplace=True)
scraped_data.head()
```

```python
#read in csv

csv_file = "Resources/url_sheet.csv"
urls = pd.read_csv(csv_file)
urls.head()
```

```python
url_merge = pd.merge(scraped_data,  
                      urls,  
                      on ='url',  
                      how ='left') 
url_merge
```

```python
store_merge = pd.merge(url_merge,  
                      stores,  
                      on ='store_name',  
                      how ='left') 
store_merge.head()
```

```python
category_merge = pd.merge(store_merge,  
                      items,  
                      on ='category_name',  
                      how ='left') 
category_merge.head()
```

```python
category_merge.rename(columns={'product_name_x':'product_name'}, inplace=True)
category_merge.rename(columns={'store_url_x':'store_url'}, inplace=True)
category_merge.rename(columns={'id_x' : 'store_id'}, inplace=True)
category_merge.rename(columns={'id_y' : 'category_id'}, inplace=True)
```

```python
product_price = category_merge[["product_name","category_id","store_id","list_price","scraped_date","on_sale","url", "availability"]]
product_price.set_index('product_name', inplace=True)
product_price
```

```python
connection_string = f"postgresql+psycopg2://{SQL_USERNAME}:{SQL_PASSWORD}@{SQL_IP}:{PORT}/{DATABASE}"
engine = create_engine(connection_string)

product_price.to_sql('product_price', engine, if_exists='append')
```
