```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import lxml
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import time
from functools import reduce
```

```python
executable_path = {"executable_path":ChromeDriverManager().install()}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
#Wilson item urls
wRac1 = "rackets/roger-federer-23-tennis-racket-158955"
wRac2 = "rackets/blade-team-tennis-racket-156242"
wRac3 = "rackets/burn-100s-v4-tennis-racket"
wBall = "balls/value/championship-extra-duty-tennis-balls-4-ball-can"
wBag = "bags/tour/small-super-tour-2-tennis-bag"
wAcc1 = "grips/pro-overgrip-3-pack-159468"
wAcc2 = "grips/overgrips/pro-overgrip-sensation-3-pack"
wShoe = "shoes/womens/women-s-rush-pro-3-0-159129"
```

```python
# masterDF = pd.DataFrame()

wProductsWeWant = [wRac1, wRac2, wRac3, wBall, wBag, wAcc1, wAcc2, wShoe]

productNames = []
listPrices = []
onSales = []
urls = []
availabilitys = []
scrapedDates = []
    

for wProductToScrape in wProductsWeWant:
    
    #get correct URL format for the products
    urlW = f"https://www.wilson.com/en-us/tennis/{wProductToScrape}"
    response = requests.get(urlW)
    
    soup = bs(response.content, "lxml")
        
    #pull data, need to find where data is located
    productName = soup.find("div", class_="product-name").text.strip()
    listPrice = soup.find("span", class_="price").text.strip()
    url1 = urlW

    if soup.find("span", class_="old-price"):
        onSale = True
    else:
        onSale = False
    
    if soup.find("span", class_="button-title").text == 'Out of Stock':
        availability=False
    else:
        availability=True
        
    scrapedDate = pd.to_datetime('today')



    #appending data
    productNames.append(productName)
    listPrices.append(listPrice)
    onSales.append(onSale)
    urls.append(url1)
    availabilitys.append(availability)
    scrapedDates.append(scrapedDate)
    
#create DF  
wDF=pd.DataFrame()
wDF["product_name"] = productNames
wDF["list_price"] = listPrices
wDF["scraped_date"] = scrapedDates
wDF["on_sale"] = onSales
wDF["url"] = urls
wDF["availability"] = availabilitys
wDF.set_index('product_name', inplace=True)


#add info to DF
#masterDF = masterDf.append(df)
```
