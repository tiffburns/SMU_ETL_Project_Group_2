```python
# import dependences 

import pandas as pd
from sqlalchemy import create_engine, inspect
import numpy as np
import datetime
import psycopg2
```

```python
#read in csv

csv_file = "Resources/url_sheet.csv"
store_df = pd.read_csv(csv_file)
store_df.head()
```

```python
stores = pd.DataFrame(store_df, columns = ['store_name', 'store_url']) 
stores
```

```python
stores_table = stores.drop_duplicates(subset=['store_url'])
stores_table.set_index('store_name', inplace=True)
stores_table
```

```python
connection_string = f"postgresql+psycopg2://{SQL_USERNAME}:{SQL_PASSWORD}@{SQL_IP}:{PORT}/{DATABASE}"
engine = create_engine(connection_string)

stores_table.to_sql('lk_store', engine, if_exists='append')
```
